Add HDI Category Column

```python
def hdi_category(h):
    if h < 0.55:
        return "Low"
    elif h < 0.70:
        return "Medium"
    elif h < 0.80:
        return "High"
    else:
        return "Very High"
```

```python
hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(hdi_category)
hdi_2022_df['HDI Category'].value_counts()
```

```python
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
```

PROBLEM 1B — HDI TREND ANALYSIS (2020–2022)

Extract Data

```python
hdi_1b = df[df['year'].isin([2020, 2021, 2022])].copy()
hdi_1b.to_csv("HDI_problem1B.csv", index=False)
```

Cleaning

```python
hdi_1b.replace(["NA", "–"], np.nan, inplace=True)
hdi_1b['hdi'] = pd.to_numeric(hdi_1b['hdi'], errors='coerce')
hdi_1b.dropna(subset=['hdi','country','year'], inplace=True)
hdi_1b.drop_duplicates(inplace=True)
```

Line Chart (5 Countries)

```python
countries = ['Nepal','India','China','United States','Germany']
subset = hdi_1b[hdi_1b['country'].isin(countries)]
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=subset, x='year', y='hdi', hue='country', marker='o')
plt.title("HDI Trend (2020–2022)")
plt.show()
```

Bar Chart — Average HDI by Region

```python
region_map = {
    # South Asia
    'Afghanistan': 'South Asia',
    'Bangladesh': 'South Asia',
    'Bhutan': 'South Asia',
    'India': 'South Asia',
    'Maldives': 'South Asia',
    'Nepal': 'South Asia',
    'Pakistan': 'South Asia',
    'Sri Lanka': 'South Asia',

    # Middle East
    'Bahrain': 'Middle East',
    'Iran': 'Middle East',
    'Iraq': 'Middle East',
    'Israel': 'Middle East',
    'Jordan': 'Middle East',
    'Kuwait': 'Middle East',
    'Lebanon': 'Middle East',
    'Oman': 'Middle East',
    'Palestine': 'Middle East',
    'Qatar': 'Middle East',
    'Saudi Arabia': 'Middle East',
    'Syria': 'Middle East',
    'United Arab Emirates': 'Middle East',
    'Yemen': 'Middle East'
}
```

```python
hdi_1b['region'] = hdi_1b['country'].map(region_map)
```

```python
hdi_1b = hdi_1b.dropna(subset=['region'])
```

```python
region_avg = (
    hdi_1b
    .groupby(['region', 'year'])['hdi']
    .mean()
    .reset_index()
)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=region_avg, x='region', y='hdi', hue='year')
plt.xticks(rotation=45)
plt.title("Average HDI by Region (2020–2022)")
plt.show()
```

Box Plot


```python
sns.boxplot(data=hdi_1b, x='year', y='hdi')
plt.title("HDI Distribution (2020–2022)")
plt.show()
```

Scatter Plot (HDI vs GNI)

```python
sns.scatterplot(data=hdi_1b, x='gross_inc_percap', y='hdi')
plt.title("HDI vs GNI per Capita")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.show()
```
